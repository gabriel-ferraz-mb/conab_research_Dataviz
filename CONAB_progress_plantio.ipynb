```python
### https://apps.fas.usda.gov/opendataweb/home
import warnings
warnings.filterwarnings('ignore')
from pandas import DataFrame
import requests
import pandas as pd
import seaborn as sns
import numpy
import datetime
import matplotlib.pyplot as plt
from datetime import time, timedelta, date
import openpyxl
plt.style.use('classic')
import psycopg2
```

```python
# Import data
conn = psycopg2.connect("host='186.202.136.178' port='5432' dbname='carbon' user='' password=''")
cur = conn.cursor()
query = "SELECT * from research.conab_base"
cur.execute(query)
colnames = [desc[0] for desc in cur.description]
df = df = DataFrame(cur.fetchall(), columns = colnames)
```

```python
df[df.select_dtypes(include=['number']).columns] *= 100
#df['data'] = pd.to_datetime(df['data'], format="%d/%b")
```
